## Methods & Results

### Summary of Methods Used

For my analysis, I began by preprocessing the `players.csv` dataset, removing unnecessary features such as `gender`, `name`, `hashedEmail`, and `played_hours`. The dataset was then split into a 75% training set and a 25% test set. I set a random seed to ensure reproducibility.

I chose **K-Nearest Neighbours (KNN)** as the classification model. During the data split, I stratified by the target variable `subscribe`. All predictor columns in the KNN recipe were standardized so they contributed equally to distance calculations.

I performed **5-fold cross-validation** on the training set to tune the number of neighbours (K). I then used the best-performing K to fit the final model, made predictions on the test set, and measured accuracy and recall.

### Data Wrangling

- I selected the relevant columns (`Age`, `experience`, `subscribe`).  
- I applied **ordinal encoding** to the `experience` variable using the order:  
  `"Amateur" < "Beginner" < "Regular" < "Pro" < "Veteran"`.  
- I converted `subscribe` to a factor.  
- I removed rows with missing values.

The recall score (‚âà0.944) showed that the model was strong at predicting the **TRUE** class (subscribers) but weak at identifying **FALSE** cases (non-subscribers).
